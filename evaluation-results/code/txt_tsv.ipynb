```python
import os
import re

def convert_txt_to_tsv(directory):
    # Directory and its subdirectories
    for root, dirs, files in os.walk(directory):
        for filename in files:
            if filename.endswith('.txt') and ('_triples' in filename or '_response' in filename):

                txt_file_path = os.path.join(root, filename)

                tsv_file_path = os.path.join(root, filename.replace('.txt', '.tsv'))

                print(f"Processing file: {txt_file_path} to create: {tsv_file_path}")

                # Open the TXT file for reading
                with open(txt_file_path, 'r', encoding='utf-8') as txt_file:
                    lines = txt_file.readlines()


                valid_triples = []

                for line in lines:
                    line = line.strip()  # Remove whitespace
                    if line:  # Ensure the line is not empty
                        try:
                            # Attempt to extract triples in various formats
                            # This regex looks for common triple structures
                            matches = re.findall(r'(\w+\(.*?\))', line)
                            if matches:
                                for match in matches:
                                    predicate, subjects_objects = match.split('(', 1)
                                    subjects_objects = subjects_objects.rstrip(')')  # Remove closing parenthesis

                                    # Split subjects and objects by commas
                                    subject_object_pairs = subjects_objects.split(',')
                                    subject = subject_object_pairs[0].strip()  # First part is the subject

                                    # Handle cases where there are multiple objects
                                    objects = subject_object_pairs[1:]  # Everything after the first part is considered as objects
                                    for obj in objects:
                                        # Clean the object string
                                        obj = obj.strip().strip('"').strip("'")
                                        if subject and obj:  # Only add valid triples
                                            tsv_line = f"{subject}\t{predicate.strip()}\t{obj}\n"
                                            valid_triples.append(tsv_line)
                            else:
                                # General structure matching
                                general_triples = re.findall(r'([^:]+)\s+(is\s+\w+)\s+(.+)', line)
                                for subject, predicate, obj in general_triples:
                                    subject = subject.strip()
                                    predicate = predicate.strip()
                                    obj = obj.strip()
                                    if subject and predicate and obj:
                                        tsv_line = f"{subject}\t{predicate}\t{obj}\n"
                                        valid_triples.append(tsv_line)

                            # Extract any meaningful information if the structured approach fails
                            generic_info = re.findall(r'(\w+)\s+([\w\s]+)\s+(.+)', line)
                            for subject, predicate, obj in generic_info:
                                subject = subject.strip()
                                predicate = predicate.strip()
                                obj = obj.strip()
                                if subject and predicate and obj:
                                    tsv_line = f"{subject}\t{predicate}\t{obj}\n"
                                    valid_triples.append(tsv_line)

                        except Exception as e:
                            print(f"Error processing line: {line}\nError: {e}")

                # Write valid triples to TSV if any were found
                if valid_triples:
                    with open(tsv_file_path, 'w', encoding='utf-8') as tsv_file:
                        tsv_file.writelines(valid_triples)  # Write all valid triples at once
                    print(f"Wrote valid triples to: {tsv_file_path}")
                else:
                    print(f"No valid entries found for: {txt_file_path}")

# Change directory depending on the generated data
directory_path = '/content/drive/My Drive/LLM-Data/llm-generated/triples_GPT-4_WB'

# Convert TXT to TSV
convert_txt_to_tsv(directory_path)
```
