```python
# Pull Information from Enslaved for Slaves that exist on Enslaved and Wikipedia

#queryv2 = f"https://lod.enslaved.org/w/api.php?action=wbsearchentities&format=json&search={filename}&language=en&type=item"
directory = "../wikitextForSlavesOnWikiandEnslaved"

# Query only returned a single result, so most likely a match. 
# Verified results are in confirmedMatches.txt
candidateMatches = open("candidateMatches.tsv", "a")

# Query returned more than one result, 
# so need to verify which one is correct manually
duplicates = open("duplicates.tsv", "a")

# Query returned no results, so need to check manually if there is a match
missing = open("missing.tsv", "a")

for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        name = filename.split(".")[0].split("(")[0].replace("_", " ")
        result = requests.get(f"https://lod.enslaved.org/w/api.php?action=wbsearchentities&format=json&search={name}&language=en&type=item").json()
        if len(result["search"]) == 1:
            match = result["search"][0]
            #print(f"{filename} matched {match['id']}")
            candidateMatches.write(f"{filename}\t{match['id']}\t{match['url']}\t{match['label']}\t{match['description']}\n")
            continue
        if len(result["search"]) == 0:
            #print(f"{filename} has no matches")
            missing.write(f"{filename}\n")
            continue
        if len(result["search"]) > 1:
            #print(f"{filename} has {len(result['search'])} duplicates")
            for i in result["search"]:                
                duplicates.write(f"{filename}\t{i['id']}\t{i['url']}\t{i['label']}\t{i['description']}\n")
        

candidateMatches.close()
duplicates.close()
missing.close()
```

```python
# Pull TTL files for confirmed matches
with open("confirmedMatches.txt", "r") as cm:
    for line in tqdm(cm.readlines()):
        line = line.strip().split("\t")
        filename = line[0].split(".")[0]
        qid = line[3]
        url = line[2]
        personName = line[4]
        
        if os.path.exists(f"./EnslavedTTL/{filename}.ttl"):
            print(f"{filename} already exists, skipping")
            continue
        else:
            with open(f"./EnslavedTTL/{filename}.ttl", "w") as t:
                ttl = requests.get(f"https://lod.enslaved.org/wiki/Special:EntityData/{qid}.ttl").text
                t.write(ttl)
                t.close()
```
